### Tratamiento de Dataframes:

1. Creacion de nuevo Dataframe con tratamiento de datos 



```python
#CREACION DE NUEVOS DATA FRAME PARA EL TRATAMIENTO DE DATOS
C_T_2023=pd.DataFrame({'Marca': pd.Series(dtype='str'),
                       'Modelo': pd.Series(dtype='str'),
                       'Año': pd.Series(dtype='str'),
                       'E.Bateria (kWh)': pd.Series(dtype='float'),
                       'Aceleración (s)':pd.Series(dtype='float'),
                       'Velocidad Max (km/h)':pd.Series(dtype='float'),
                       'Rango (km)':pd.Series(dtype='float'),
                       'Eficiencia (Wh/km)':pd.Series(dtype='float'),
                       'Vel. Carga R':pd.Series(dtype='float'),
                       'Transmision':pd.Series(dtype='str'),
                       'Num. Asientos':pd.Series(dtype='int'),
                       'Precio (GER)(EUR)':pd.Series(dtype='float'),
                       'Precio (UK)(GBL)':pd.Series(dtype='float')})


Q_T_2021=pd.DataFrame({'Marca': pd.Series(dtype='str'),
                       'Modelo': pd.Series(dtype='str'),
                       'E.Bateria (kWh)': pd.Series(dtype='float'),
                       'Aceleración (s)':pd.Series(dtype='float'),
                       'Velocidad Max (km/h)':pd.Series(dtype='float'),
                       'Rango (km)':pd.Series(dtype='float'),
                       'Eficiencia (Wh/km)':pd.Series(dtype='float'),
                       'Vel. Carga R':pd.Series(dtype='float'),
                       'Transmision':pd.Series(dtype='str'),
                       'Num. Asientos':pd.Series(dtype='int'),
                       'Precio (GER)(EUR)':pd.Series(dtype='float'),
                       'Precio (UK)(GBL)':pd.Series(dtype='float')})



```

2. Obtención de Modelo y Marca, separación por fila de la comula "Name"

```python
# se utiliza el metodo split para separar por espacios el string y obtener los string de interes
C_T_2023['Marca']=CEV2023['Name'].str.split(' ').str[0]

Q_T_2021['Marca']=QEV2021['Name'].str.split(' ').str[0]

C_T_2023['Modelo']=CEV2023['Name'].str.split(' ',1).str[1]

Q_T_2021['Modelo']=QEV2021['Name'].str.split(' ',1).str[1]

C_T_2023['Año']=CEV2023['Subtitle'].str.split(' ').str[-1]
```

3.  Obtención de valores númericos, eliminando cualquier rastro de un valor tipo 'str'

```python
# se valida  cada columna del dataframe creado para el tratamiento de datos
C_T_2023['E.Bateria (kWh)']=CEV2023['Subtitle'].str.split(' ').str[0].astype('float')
C_T_2023['Aceleración (s)']=CEV2023['Acceleration'].str.split(' ').str[0].astype('float')
C_T_2023['Velocidad Max (km/h)']=CEV2023['TopSpeed'].str.split(' ').str[0].astype('float')
C_T_2023['Rango (km)']=CEV2023['Range'].str.split(' ').str[0].astype('float')
C_T_2023['Eficiencia (Wh/km)']=CEV2023['Efficiency'].str.split(' ').str[0].astype('float')
C_T_2023['Num. Asientos']=CEV2023['NumberofSeats'].astype('int')
C_T_2023['Transmision']=CEV2023['Drive'].astype('str')

Q_T_2021['E.Bateria (kWh)']=QEV2021['Subtitle'].str.split(' ').str[10].astype('float')
Q_T_2021['Aceleración (s)']=QEV2021['Acceleration'].str.split(' ').str[0].astype('float')
Q_T_2021['Velocidad Max (km/h)']=QEV2021['TopSpeed'].str.split(' ').str[0].astype('float')
Q_T_2021['Rango (km)']=QEV2021['Range'].str.split(' ').str[0].astype('float')
Q_T_2021['Eficiencia (Wh/km)']=QEV2021['Efficiency'].str.split(' ').str[0].astype('float')
Q_T_2021['Num. Asientos']=QEV2021['NumberofSeats'].astype('int')
Q_T_2021['Transmision']=QEV2021['Drive'].astype('str')

```

4. Las columnas de Velocidad de carga rápida, precio en Alemania y el Reino Unido tienen valores faltantes, por lo tanto se buscara en
internet desde la siguiente fuente:
    
    https://ev-database.org/

     Sin embargo, es importante primero reconocer cuales tienen valores faltantes, para lo cual se asignaran en el Dataframe tratado, tal como esta para que luego del completado de valores se le asigne su tipo de valor:




```python
C_T_2023['Vel. Carga R'] = CEV2023['FastChargeSpeed'].str.split(' ').str[0]

miss_battery_c = C_T_2023.loc[C_T_2023['Vel. Carga R'].str.isnumeric(
) == False, 'Modelo']

display("Carga Rapida faltante en C_T_2023")

display(miss_battery_c)

Q_T_2021['Vel. Carga R'] = QEV2021['FastChargeSpeed'].str.split(' ').str[0]

miss_battery_q = Q_T_2021.loc[Q_T_2021['Vel. Carga R'].str.isnumeric(
) == False, 'Modelo']

display("Carga Rapida faltante en Q_T_2021")

display(miss_battery_q)

# Asignacion de datos obtenidos desde la funete antes mencionada:

ind_b_c = miss_battery_c.index
ind_b_q = miss_battery_q.index

# Obteniendo del internet para C_T_2023 y Q_T_2021 =

miss_batteries = [23, 22, 110, 110, 32]


# Asignacion en C_T_2023
for i in range(0, 3):

    C_T_2023.iloc[[ind_b_c[i]], [7]] = miss_batteries[i]

# Asignacion en Q_T_2021

for i in range(0, 5):

    C_T_2023.iloc[[ind_b_q[i]], [7]] = miss_batteries[i]
```

Para las columnas del precio del vehículo, ya sea de UK o Alemania primero se obtendrá los modelos que carecen de información en ambos países, en esos casos se buscara en internet su valor actual de mercado.

En caso de no poseer el precio en solo uno de los dos países se ejecutará el cambio de la moneda de euro a libras y viceversa dado sea el caso:


```python
# obtencion de datos dataframe inicial:

C_T_2023[['Precio (GER)(EUR)', 'Precio (UK)(GBL)']
         ] = CEV2023[['PriceinGermany', 'PriceinUK']]

Q_T_2021[['Precio (GER)(EUR)', 'Precio (UK)(GBL)']
         ] = QEV2021[['PriceinGermany', 'PriceinUK']]

# Reemplazo coma por punto:

C_T_2023['Precio (GER)(EUR)'] = C_T_2023['Precio (GER)(EUR)'].str.replace(
    ',', '')
C_T_2023['Precio (UK)(GBL)'] = C_T_2023['Precio (UK)(GBL)'].str.replace(
    ',', '')

Q_T_2021['Precio (GER)(EUR)'] = Q_T_2021['Precio (GER)(EUR)'].str.replace(
    ',', '')
Q_T_2021['Precio (UK)(GBL)'] = Q_T_2021['Precio (UK)(GBL)'].str.replace(
    ',', '')

# Eliminación de símbolo del dinero:

C_T_2023['Precio (GER)(EUR)'] = C_T_2023['Precio (GER)(EUR)'].str.extract(
    '(\d+)', expand=False)
C_T_2023['Precio (UK)(GBL)'] = C_T_2023['Precio (UK)(GBL)'].str.extract(
    '(\d+)', expand=False)

Q_T_2021['Precio (GER)(EUR)'] = Q_T_2021['Precio (GER)(EUR)'].str.extract(
    '(\d+)', expand=False)
Q_T_2021['Precio (UK)(GBL)'] = Q_T_2021['Precio (UK)(GBL)'].str.extract(
    '(\d+)', expand=False)

# Obtención de Modelos con ambos precios faltantes en ambos dataframes:

miss_all_prices_c = C_T_2023.loc[(C_T_2023['Precio (GER)(EUR)'].isna()) & (
    C_T_2023['Precio (UK)(GBL)'].isna()), 'Modelo']

display("Ambos precios Faltantes en C_T_2023")
display(miss_all_prices_c)

ind_c = miss_all_prices_c.index


miss_all_prices_q = Q_T_2021.loc[(C_T_2023['Precio (GER)(EUR)'].isna()) & (
    Q_T_2021['Precio (UK)(GBL)'].isna()), 'Modelo']

display("Ambos precios Faltantes en Q_T_2021")
display(miss_all_prices_q)

ind_q = miss_all_prices_q.index


# Obteniendo del internet para C_T_2023 y Q_T_2021 =

l_prices_C_Eur = [44700, 46700, 29900, 70626, 58895, 68595, 60095, 69795,
                  30990, 83590, 94590, 37995, 51188, 61188, 71188, 59188, 63188, 73188]

l_prices_Q_Eur=[53668,43900,33133,36400,50880,50880]

#Asignacion en C_T_2023
for i in range(0, len(l_prices_C_Eur)):

    C_T_2023.iloc[[ind_c[i]], [11]] = l_prices_C_Eur[i]

#Asignacion en Q_T_2021

for i in range(0, len(l_prices_Q_Eur)):

    Q_T_2021.iloc[[ind_q[i]], [11]] = l_prices_Q_Eur[i]


```

En caso de faltar su precio en (EUR), se calcula su equivalencia a partir del cambio con la moneda (GBP) y de igual forma en viceversa::

```python
#Multiplicacion x1.16 para transformar de euro a gbl y por 0.86 para la transformacion inversa
C_T_2023['Precio (GER)(EUR)'] = C_T_2023['Precio (GER)(EUR)'].fillna(C_T_2023['Precio (UK)(GBL)'].astype(float)*1.16)
C_T_2023['Precio (UK)(GBL)'] = C_T_2023['Precio (UK)(GBL)'].fillna(C_T_2023['Precio (GER)(EUR)'].astype(float)*0.86)

Q_T_2021['Precio (GER)(EUR)'] = Q_T_2021['Precio (GER)(EUR)'].fillna(Q_T_2021['Precio (UK)(GBL)'].astype(float)*1.16)
Q_T_2021['Precio (UK)(GBL)'] = Q_T_2021['Precio (UK)(GBL)'].fillna(Q_T_2021['Precio (GER)(EUR)'].astype(float)*0.86)

C_T_2023['Precio (GER)(EUR)']=C_T_2023['Precio (GER)(EUR)'].astype(float)
C_T_2023['Precio (UK)(GBL)']=C_T_2023['Precio (UK)(GBL)'].astype(float)

Q_T_2021['Precio (GER)(EUR)']=Q_T_2021['Precio (GER)(EUR)'].astype(float)
Q_T_2021['Precio (UK)(GBL)']=Q_T_2021['Precio (UK)(GBL)'].astype(float)

```

Visualización Final de Data Frame:

```python
print("Vehiculos eléctricos del 2023")
display(C_T_2023.sample(3))
print("Vehiculos eléctricos  del 2021")
display(Q_T_2021.sample(3))


```

### Estadística Descriptiva:

#### Agrupación por marca:

Obtención y valoración de promedio del precio:

```python
#Agrupacion por el metodo group by:
# se realiza la operacion de obtencion de media

Label_P_C=C_T_2023.groupby(['Marca'], as_index=False).agg({'Precio (GER)(EUR)':['mean']})

display('Promedio de precios por marca en C_T_2023')
display(Label_P_C.head(3))

Label_P_Q=Q_T_2021.groupby(['Marca'], as_index=False).agg({'Precio (GER)(EUR)':['mean']})
display('Promedio de precios por marca en Q_T_2021')
display(Label_P_Q.head(3))


```

### Graficación

```python
# GRAFICACION DE RESULTADOS

x_data_c=Label_P_C['Marca']
y_data_c=Label_P_C['Precio (GER)(EUR)']

x_data_q=Label_P_Q['Marca']
y_data_q=Label_P_Q['Precio (GER)(EUR)']


# se declara figura y se brinda todos los detalles para mejorar la presentacion de la misma, esto se realiza en cada proceso de graficación
fig, axis = plt.subplots(1,1,figsize=(12, 4))

axis.plot(x_data_c,y_data_c,'bx-',label='Mas baratos 2023',linewidth=1.5,alpha=0.75)

axis.plot(x_data_q,y_data_q,'g*',label='Mas rápidos 2021',linewidth=1.5,alpha=0.75)

axis.legend(loc='upper left')

axis.set_title('Comparación de precio:  2021 vs 2023')

axis.set_xlabel('Marca')

axis.set_ylabel('Precio ')

plt.xticks(rotation = 90)
plt.show
```

Como se puede observar en la gráfica el vehículo con el promedio de precios mas alto es el Roll Royce, mientras que el mas barato se encuentra entre los vehículos de marcas: Smart, Seat, Sono y Diaca


Modelos de auto con el precio minimo entre todas las marcas:

```python
# se desarrolla metodo groupby con la operacion min para obtener el precio minimo de cada grupo dado por la marca
Min_C=C_T_2023.groupby(['Marca','Modelo'], as_index=True).agg({'Precio (GER)(EUR)':['min']})

MIN_C=Min_C.idxmin() #método  para obtener indice mínimo 

display("En C_T_2023")
display(MIN_C)

Min_Q=Q_T_2021.groupby(['Marca','Modelo'], as_index=True).agg({'Precio (GER)(EUR)':['min']})

MIN_Q=Min_Q.idxmin() #método  para obtener indice máximo 

display("En Q_T_2021")
display(MIN_Q)


```

Del 2021 al 2023 el Smart, EQ fortwo coupe sigue siendo uno de los vehículos más rápidos y asequibles del mercado


Modelos de auto con el precio mas alto entre todas las marcas:

```python
# se desarrolla metodo groupby con la operacion max para obtener el precio máximo de cada grupo dado por la marca
Max_C=C_T_2023.groupby(['Marca','Modelo'], as_index=True).agg({'Precio (GER)(EUR)':['max']})
 
# metodo para obtener el índice máximo 
MAX_C=Max_C.idxmax()

display("En C_T_2023")
display(MAX_C)

Max_Q=Q_T_2021.groupby(['Marca','Modelo'], as_index=True).agg({'Precio (GER)(EUR)':['max']})

MAX_Q=Max_Q.idxmax()

display("En Q_T_2021")
display(MAX_Q)
```
